```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
reference = np.concatenate([cv2.imread(r'C:\Users\APrap\Documents\CREATE\Pick-and-Place\Pictures\Utils\X0_template.png'), cv2.imread(r'C:\Users\APrap\Documents\CREATE\Pick-and-Place\Pictures\Utils\X1_template.png')], axis=1)
plt.imshow(reference)

def check_tip(image):
    res = cv2.matchTemplate(reference, image[200:-200, 100:-100], cv2.TM_SQDIFF_NORMED)
    
    plt.imshow(res)
    
    if np.argmin(res) < reference.shape[1]/4:
        print('No sample')
    else:
        print('Sample')
    
```

```python
image = cv2.imread(r'C:\Users\APrap\Documents\CREATE\Pick-and-Place\Pictures\macro\macro_image_278.png')
check_tip(image)
```
