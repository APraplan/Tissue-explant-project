```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math

import os
```

```python
frame = cv2.imread('..\Pictures\ArUco.PNG')
if frame is None:
    print("Error loading image")
```

```python
dictionary = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_4X4_50)
parameters =  cv2.aruco.DetectorParameters()
detector = cv2.aruco.ArucoDetector(dictionary, parameters)   
markerCorners, markerIds, _ = detector.detectMarkers(frame)
```

```python
selected_id = 1
try:
    id = np.where(markerIds == selected_id)[0][0]
except:
    id = None
```

```python
center_point = np.mean(markerCorners[id][0], axis=0)
imshow = cv2.circle(frame.copy(), (int(center_point[0]), int(center_point[1])), 8, (255, 0, 0), -1)
```

```python
for i in range(len(markerCorners[id][0])):
    print('x coordinate ', markerCorners[id][0][i][0], ' y coodinate ', markerCorners[id][0][i][1])
    imshow = cv2.circle(imshow, (int(markerCorners[id][0][i][0]), int(markerCorners[id][0][i][1])), 8, (0, 0, 255), -1)
    
```

```python
plt.imshow(imshow)
```
