```python
dictionary = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_4X4_50)
parameters =  cv2.aruco.DetectorParameters()
detector = cv2.aruco.ArucoDetector(dictionary, parameters)
```

```python
def draw(markerCorners, img):
        
    int_corners = np.int32(markerCorners)

    font = cv2.FONT_HERSHEY_SIMPLEX
    fontScale = 2
    color = (255,0, 0)
    thickness = 4

    out = cv2.polylines(img,  [int_corners], True, (0, 255, 0), 4)
    for i in range(len(int_corners)):
        out = cv2.putText(out, str(i+1), int_corners[i], font, 
                    fontScale, color, thickness, cv2.LINE_AA)
    
    return out 
```

```python
img = img1.copy()
markerCorners, markerIds, rejectedCandidates = detector.detectMarkers(img)
# print(len(markerCorners[0][0]))
markerCorners = np.array(markerCorners[0][0])

perimeter = cv2.arcLength(markerCorners, True)
pixel_cm_ratio = perimeter / 400
# print(perimeter)

img = draw(markerCorners, img)
center = (img.shape[1]//2, img.shape[0]//2)
print('center ', center)
print('corners ', markerCorners)

print(markerCorners[0][0]-center[0], markerCorners[0][1]-center[1])
   
    
plt.imshow(img)
```
